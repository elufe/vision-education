```python
import cv2
import numpy as np

def backproject(source, target, levels = 2, scale = 1):
        hsv = cv2.cvtColor(source,  cv2.COLOR_BGR2HSV)
        hsvt = cv2.cvtColor(target, cv2.COLOR_BGR2HSV)
        # calculating object histogram
        roihist = cv2.calcHist([hsv],[0, 1], None, [levels, levels], [0, 180, 0, 256] )

        # normalize histogram and apply backprojection
        cv2.normalize(roihist,roihist,0,255,cv2.NORM_MINMAX)
        dst = cv2.calcBackProject([hsvt],[0,1],roihist,[0,180,0,256], scale)
        return dst
    

img = cv2.imread('menu.jpg')
img = cv2.resize(img, (500,600))
img1 = img.copy()

cv2.pyrMeanShiftFiltering(img1, 2, 10, img1, 4)


backproj = np.uint8(backproject(img1, img1, levels = 2))
backproj = 255 - backproj
backproj = cv2.morphologyEx(backproj, cv2.MORPH_CLOSE, (5,5))

cv2.imshow('back', backproj)

kernel = np.ones((5,5),np.uint8)
img_bin_bound = cv2.threshold(backproj, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
img_bin_bound = cv2.morphologyEx(img_bin_bound, cv2.MORPH_CLOSE, kernel)

cv2.Canny(img_bin_bound, 100, 200)
_,contours, hierarchy = cv2.findContours(img_bin_bound,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

contours = sorted(contours, key = cv2.contourArea)

x,y,w,h = cv2.boundingRect(contours[-1])




img2 = img.copy()
img2 = img2[y:y+h,x:x+w]

cv2.pyrMeanShiftFiltering(img2, 2, 10, img2, 4)
#cv2.medianBlur(img2, 3)

img_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

img_bin = cv2.threshold(img_gray, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

img_bin = cv2.morphologyEx(img_bin, cv2.MORPH_CLOSE, kernel)

cv2.Canny(img_bin, 100, 200)

_,contours, hierarchy = cv2.findContours(img_bin,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    if cv2.arcLength(cnt, True) < 30:
        continue
    rc = cv2.boundingRect(cnt)
    cv2.rectangle(img, (x+rc[0], y+rc[1]), (x+rc[0]+rc[2], y+rc[1]+rc[3]), (255,0,0),2)


cv2.imshow('img',img)
cv2.imshow('gray',img_gray)
cv2.imshow('bin', img_bin)
cv2.waitKey()
cv2.destroyAllWindows()
```
