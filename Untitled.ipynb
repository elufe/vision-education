```python
import cv2

src1 = cv2.imread('man_chromakey.jpg')
src2 = cv2.imread('street.jpg')

#연산에 편의성을 위한 resize
src1 = cv2.resize(src1, (300,400))
src2 = cv2.resize(src2, (640,400))

#인물이 들어갈 배경의 위치
src3 = src2[:,170:470]

#mask 생성
src_hsv = cv2.cvtColor(src1, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(src_hsv, (50, 150, 0), (80, 255, 255))

mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)

#배경에서 인물이 들어갈 위치 0으로 초기화
src3 = cv2.bitwise_and(src3, mask)

# mask 반전
mask = cv2.bitwise_not(mask)
# 인물사진에서 배경제거
src1 = cv2.bitwise_and(src1, mask)

# 배경에 인물 합성
src3 = cv2.bitwise_or(src1, src3)
src2[:,170:470] = src3

cv2.imshow('src1',src1)
cv2.imshow('src2',src2)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2

# 동영상 불러오기
src = cv2.VideoCapture('chroma_key1.mov')
src2 = cv2.VideoCapture('chroma_key2.mov')

# 동영상 가로, 세로, fps
w = round(src.get(cv2.CAP_PROP_FRAME_WIDTH))
h = round(src.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = round(src2.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc('M','J','P','G')
out = cv2.VideoWriter("200711.avi", fourcc, fps, (w,h))

while True:
    # 동영상에서 프레임 읽기
    ret1, cap = src.read()
    ret2, cap2 = src2.read()
    
    # 동영상이 끝났는지 확인
    if ret1 == False or ret2 == False:
        break

    # mask 생성
    cap_hsv = cv2.cvtColor(cap, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(cap_hsv, (50, 150, 0), (80, 255, 255))
    mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    
    # 두번째 영상에서 첫번째 영상이 들어갈 위치 0으로 초기화
    cap2 = cv2.bitwise_and(cap2, mask)
    
    # mask 반전
    mask = cv2.bitwise_not(mask)
    
    # 첫번째 영상 배경 제거
    cap = cv2.bitwise_and(cap, mask)
    
    # 영상 합성
    cap2 = cv2.bitwise_or(cap, cap2)
        
    out.write(cap2)
    
    cv2.imshow("frame", cap2)
    
    if cv2.waitKey(10) == 27:
        break
    
src.release()
out.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

src = cv2.imread('lena.jpg')

aff = np.array([[1,0,20],
               [0,1,10]], dtype = np.float32)

dst = cv2.warpAffine(src, aff, (0,0))
dst2 = cv2.warpAffine(src, aff, (0,0), None, cv2.INTER_LINEAR, cv2.BORDER_CONSTANT, (255,0,0))

cv2.imshow('src',src)
cv2.imshow('dst',dst)
cv2.imshow('dst2',dst2)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

src = cv2.imread('LJ.jpg')

rc1 = (380, 400) 
rc2 = (300, 300)
cpy = src.copy()
cv2.rectangle(cpy, rc1, rc2, (0 , 0, 255), 2)
cv2.imshow('src',cpy)
cv2.waitKey()

for i in range(1,4):
    src = cv2.pyrDown(src)
    cpy = src.copy()
    cv2.rectangle(cpy, rc1, rc2, (0, 0, 255), 2, shift = i)
    cv2.imshow('src',cpy)
    cv2.waitKey()
    
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('aaa.jpg')
rows, cols, _ = img.shape

pst1 = np.float32([[100,50],[200,50],[100,200]])
pst2 = np.float32([[80,70],[210,60],[250,120]])

mat = cv2.getAffineTransform(pst1, pst2)
dst = cv2.warpAffine(img, mat, (int(cols*1.5),rows))

cv2.imshow('img',img)
cv2.imshow('dst',dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('b.jpg')
img = cv2.GaussianBlur(img, (5,5), cv2.BORDER_DEFAULT)

img1 = img.copy()
img1 = cv2.Canny(img1, 100,300)

img1 = cv2.bitwise_not(img1)
img1 = cv2.cvtColor(img1, cv2.COLOR_GRAY2BGR)

img = cv2.bitwise_and(img, img1)
cv2.imshow('img',img)
cv2.imshow('img1', img1)


cv2.waitKey()
cv2.destroyAllWindows()
```
